```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("DemoSparkApp").getOrCreate()
```

```python
data = [1,3,2,5,57]
```

```python
rdd = spark.sparkContext.parallelize(data)
```

```python
for i in rdd.collect():
    print(i)
```

```python
spark.stop()
```
